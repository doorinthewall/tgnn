```python
import networkx as nx
import seaborn as sns
import numpy as np
import os
import json
```

House

```python
path = '/home/Container/bgnn/datasets/house/'

graph = nx.read_graphml(os.path.join(path, 'graph.graphml'))

graph = nx.relabel_nodes(graph, {str(i): i for i in range(len(graph))})

nx.number_connected_components(graph)
```

```python
seed = '0'
with open(os.path.join(path, 'masks.json'), 'rb') as fp: 
    masks = json.load(fp)
    train_mask, val_mask, test_mask = masks[seed]['train'], masks[seed]['val'], masks[seed]['test']
```

```python
subgrpah = graph.subgraph(train_mask)
```

```python
nx.number_connected_components(subgrpah)
```

```python
sns.set()
```

```python
sns.kdeplot(list(map(lambda x: np.log(len(x)), nx.connected_components(graph))), fill=True)
sns.kdeplot(list(map(lambda x: np.log(len(x)), nx.connected_components(subgrpah))), fill=True)
```

```python
degrees = [d for n, d in graph.degree()]
degrees_ = [d for n, d in subgrpah.degree()]

sns.distplot(degrees)
sns.distplot(degrees_)
```

Degrees

```python
def get_graph(path):
    graph = nx.read_graphml(os.path.join(path, 'graph.graphml'))
    graph = nx.relabel_nodes(graph, {str(i): i for i in range(len(graph))})
    return graph
```

```python
path = '/home/Container/bgnn/datasets/house/'
graph = get_graph(path)
house_degrees = [d for n, d in graph.degree()]

path = '/home/Container/bgnn/datasets/county/'
graph = get_graph(path)
county_degrees = [d for n, d in graph.degree()]

path = '/home/Container/bgnn/datasets/vk/'
graph = get_graph(path)
vk_degrees = [d for n, d in graph.degree()]
```

```python
import matplotlib.pyplot as plt
```

```python
sns.kdeplot(np.log1p(house_degrees), label='House', fill=True)
sns.kdeplot(np.log1p(county_degrees), label='County', fill=True)
sns.kdeplot(np.log1p(vk_degrees), label='Vk', fill=True)
plt.legend()
plt.xlim(0,5)
plt.xlabel('Degrees')
plt.savefig('/home/Container/pics/Degrees.pdf', format='pdf', bbox_inches='tight')
```
